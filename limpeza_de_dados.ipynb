```python
import pandas as pd
import numpy as np
```

## Normalização dos dados

```python
df_customer = pd.json_normalize(df.customer)
df_customer.head(5)
```

```python
df_phone = pd.json_normalize(df.phone)
df_phone.head(5)
```

```python
df_internet = pd.json_normalize(df.internet)
df_internet.head(5)
```

```python
df_account = pd.json_normalize(df.account)
df_account.head(5)
```

## Unindo as tabelas

```python
dados = df[list(df.columns[:2])]
dados.head(5)
```

```python
tabelas = [df_customer, df_phone, df_internet, df_account]

for tabela in tabelas:
    concat_df = tabela
    dados = pd.concat([dados, concat_df], axis = 1)

dados.head(5)
```

## Sumarização do dado

```python
dados.info()
```

```python
for column in dados.columns:
    print(f'---- {column} ----')
    print(dados[column].value_counts())
    print('')
```

#### A coluna 'Churn' apresenta 224 valores vazios que devem ser apagados

```python
dados = dados[dados['Churn'] != '']
dados.reset_index(drop = True, inplace = True)
```

```python
for column in dados.columns:
    print(f'---- {column} ----')
    print(dados[column].value_counts())
    print('')
```
